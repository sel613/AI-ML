```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
#import dataset
df = pd.read_csv("SummerOlympicRecords.csv")

print("According to Given Data Summer Olympics have held in ",len(df["City"].unique())," cities.")
```

## 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
GoldMedals = {}
for i in df["Sport"].unique():
    Fm = pd.DataFrame(df[df["Sport"] == i])
    GoldMedals[i] = len(Fm[Fm["Medal"] == "Gold"])

Sports = list(GoldMedals.keys())
Sports.sort(reverse = True)
NoOfGolds = list(GoldMedals.values())
NoOfGolds.sort(reverse = True)

# plot
plt.figure(figsize = (18, 5))
plt.bar(Sports[:], NoOfGolds[:], color = "blue", width = 0.2)
plt.title("Sports With No of Gold Medals")
plt.xlabel("Sports")
plt.xticks(rotation='vertical')
plt.ylabel("No Of Gold Medals")
plt.show()

plt.figure(figsize = (10, 5))
plt.bar(Sports[:5], NoOfGolds[:5], color = "green", width = 0.5)
plt.title("Sports With High Gold Medals (Top 5)")
plt.xlabel("Sports")
plt.ylabel("No Of Gold Medals")
plt.show()
```

## 3. Which sport is having most number of medals so far? (Top 5)

```python
GoldMedals = {}
for i in df["Sport"].unique():
    Fm = pd.DataFrame(df[df["Sport"] == i])
    GoldMedals[i] = len(Fm["Medal"] )

Sports = list(GoldMedals.keys())
Sports.sort(reverse = True)
NoOfGolds = list(GoldMedals.values())
NoOfGolds.sort(reverse = True)

# plot

plt.figure(figsize = (18, 5))
plt.bar(Sports[:], NoOfGolds[:], color = "red", width = 0.2)
plt.title("Sports & No of Medals in it.")
plt.xlabel("Sports")
plt.xticks(rotation='vertical')
plt.ylabel("No Of Medals")
plt.show()

plt.figure(figsize = (10, 5))
plt.bar(Sports[:5], NoOfGolds[:5], color = "yellow", width = 0.5)
plt.title("Sports With High Medals (Top 5)")
plt.xlabel("Sports")
plt.ylabel("No Of Medals")
plt.show()
```

## 4. Which player has won most number of medals? (Top 5)
